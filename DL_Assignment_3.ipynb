```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv', na_values=['N/A'])
df.head()
```

```python
df.drop(columns=['id'], inplace = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
bmi_median = df['bmi'].median()
df['bmi'] = df['bmi'].fillna(bmi_median)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
print(df['gender'].unique())
```

```python
cat_cols = ['gender', 'work_type', 'smoking_status']
binary_cat_cols = ['ever_married', 'Residence_type']

le = LabelEncoder()
for col in binary_cat_cols:
    df[col] = le.fit_transform(df[col])

df = pd.get_dummies(df, columns = cat_cols, dtype = int)
df.head()
```

```python
target_col = "stroke"
X = df.drop(columns=[target_col])
y = df[target_col]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

scale_cols = ['age', 'avg_glucose_level', 'bmi']

scaler_sd = StandardScaler()
X_train[scale_cols] = scaler_sd.fit_transform(X_train[scale_cols])
X_test[scale_cols] = scaler_sd.transform(X_test[scale_cols])
```

```python
X_train_tensor = torch.from_numpy(X_train.values)
X_test_tensor = torch.from_numpy(X_test.values)

X_train_tensor = torch.tensor(X_train_tensor, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_tensor, dtype=torch.float32)

y_train_tensor = torch.from_numpy(y_train.values)
y_test_tensor = torch.from_numpy(y_test.values)

y_train_tensor = torch.tensor(y_train_tensor, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_tensor, dtype=torch.float32)
```

```python
X_train_tensor.shape[1]
```

## Preprocessing Decisions

1. I remove id feature because it has no importance in model training. Here, I got null values in bmi feature. So I replace the null values with the median value of this feature, which is outlier-resistant.

2. I got binary categorial features like ever_married, Residence_type, where I implement Label encoding and implement one-hot-encoding to other categorial features to convert them into numerical values, which will be helpful in model training.

3. Then I do train test split and implement standard scaling in age, avg_glucose_level, and bmi because this feature got bigger values than other feature which will create bias in the model without scaling.

```python
class Model(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, 1),
            nn.Sigmoid()
        )

    def forward(self, features):
      out = self.network(features)
      return out
```

```python
torch.manual_seed(42)
num_features = X_train_tensor.shape[1]
model = Model(num_features)
print(model)
```

### Explanation:

Input Layer: 18 neurons

Hidden Layer 1: 16 neurons with ReLU activation function

Hidden Layer 2: 8 neurons with ReLU activation function

Output Layer: 1 neuron with Sigmoid function

Here, we use ReLU in hidden layers to avoid vanishing gradients and let the model learn non-linear relationships. And it is a binary classification task which is why we choose Sigmoid for the output.

```python
from torchinfo import summary
summary(model, input_size=(X_train_tensor.shape[0], num_features))
```

```python
learning_rate = 0.001
loss_fn = nn.BCELoss()

optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

### Justification:

Here I use Binary cross entorphy loss because it is a binary classification task. Adam is a generalized and improved version of plain SGD, which is why I chose this optimizer for my dataset.

```python
epochs = 100

for epoch in range(epochs):
    y_pred = model(X_train_tensor)
    loss = loss_fn(y_pred, y_train_tensor.reshape(-1,1))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
      print(f"Epoch: {epoch+1}, Loss: {loss.item()}")
```

### Pipeline Explanation:

At first, take the learning rate and the epoch number. Run loop on epochs

2. Forward pass
3. Calculate loss
4. Clear old gradients
5. Backward pass
6. Lastly, update weights

```python
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.3).float()
    accuracy = (y_pred == y_test_tensor).float().mean()
    print(f"Accuracy: {accuracy.item()}")
```

```python
with torch.no_grad():
    y_pred = model(X_train_tensor)
    y_pred = (y_pred > 0.3).float()
    accuracy = (y_pred == y_train_tensor).float().mean()
    print(f"Accuracy: {accuracy.item()}")
```

### Explanation:

Here, we have train data accuracy and the test data accuracy very close. No big difference between the accuracy numbers. So that we can say the model is not underfit or overfit. This model works really well.

```python
epochs = 300

for epoch in range(epochs):
    y_pred = model(X_train_tensor)
    loss = loss_fn(y_pred, y_train_tensor.reshape(-1,1))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1)<=10:
      print(f"Epoch: {epoch+1}, Loss: {loss.item()}")
```

```python
with torch.no_grad():
    y_pred_update = model(X_test_tensor)
    y_pred_update = (y_pred_update > 0.3).float()
    accuracy = (y_pred_update == y_test_tensor).float().mean()
    print(f"Accuracy: {accuracy.item()}")
```

### Explanation:

After adding more epochs, we can see that the loss of the model has decreased, and the accuracy is improving. So now we can say the model is making better predictions. But in this dataset, there is a severe class imbalance, and after many epochs, there is a potential risk of overfitting also.

### Questions:

1. Why must gradients be reset every epoch?

2. What happens if learning rate is too high?

3. What happens if learning rate is too small?

4. Why do we define layers inside the constructor (init) and not inside forward()?

Answer 1:

If gradients are not reset in every epoch, then the gradients from the previous epoch add up to the current epoch. As a result, it will give wrong weight updates, and the model will learn incorrectly.

Answer 2:

If the learning rate is too high then the model updates weights too aggresively. In this case, there is a huge possibility model will jump over the optimal solution.

Answer 3:

If the learning rate is too low, then the model update weights too slowly. Thereafter,  the model needs thousands of epochs to experience meaningful improvement.

Answer 4:

If layers are defined in forward(), new weights are created every time whenever the forward pass runs. As a result, model can never learn. On the other hand, If layers are defined inside the constructor (init) then the weights are created once and reuses them in every forward pass so the model learns well.
